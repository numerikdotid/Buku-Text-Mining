```python
auth = tweepy.OAuthHandler(consumer_key, consumer_key_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
tweets_list = tweepy.Cursor(api.search, q='#mypertamina -filter:retweets', tweet_mode='extended',lang="id", since='2022-07-05', until='2022-07-16').items(1000)
```

```python
output = []
for tweet in tweets_list:
    text = tweet._json["full_text"]
    print(text)
    user = tweet.user.name
    created = tweet.created_at
    favourite_count = tweet.favorite_count
    retweet_count = tweet.retweet_count

    line = {'Created' : created,'Content' : text, 'User' : user, 'Favorite Count' : favourite_count, 'Retweet Count' : retweet_count}
    output.append(line)
```

```python
import pandas as pd

df = pd.DataFrame(output)
df.to_csv('mypertamina.csv', mode='a', header=False)
```
